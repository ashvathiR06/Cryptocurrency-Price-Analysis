```python
# prompt: cryptocurrency price analysis with coding

!pip install yfinance pandas matplotlib

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the cryptocurrency ticker symbol (e.g., BTC-USD for Bitcoin)
ticker = "BTC-USD"

# Download historical data (adjust the period as needed)
data = yf.download(ticker, period="1y")

# Display basic statistics
print(data.describe())

# Plot the closing price
plt.figure(figsize=(12, 6))
plt.plot(data["Close"])
plt.title(f"{ticker} Closing Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()

# Calculate and plot moving averages (e.g., 50-day and 200-day)
data["MA50"] = data["Close"].rolling(window=50).mean()
data["MA200"] = data["Close"].rolling(window=200).mean()

plt.figure(figsize=(12, 6))
plt.plot(data["Close"], label="Close")
plt.plot(data["MA50"], label="MA50")
plt.plot(data["MA200"], label="MA200")
plt.title(f"{ticker} Closing Price with Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()


# Calculate daily returns
data["Daily_Return"] = data["Close"].pct_change()

# Plot daily returns
plt.figure(figsize=(12, 6))
plt.plot(data["Daily_Return"])
plt.title(f"{ticker} Daily Returns")
plt.xlabel("Date")
plt.ylabel("Percentage Change")
plt.grid(True)
plt.show()


#Further analysis (example: Volatility)
#Calculate the rolling standard deviation (volatility)
data['Volatility'] = data['Close'].rolling(window=20).std()  #20-day rolling volatility
plt.figure(figsize=(12, 6))
plt.plot(data['Volatility'])
plt.title(f"{ticker} Volatility (20-day Rolling Std Dev)")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.grid(True)
plt.show()

```
